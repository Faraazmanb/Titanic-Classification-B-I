```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pdf=pd.read_csv("C:\\Users\\ELCOT\\Desktop\\titanic\\Titanic-Dataset.csv")
```

```python
dataset=pdf
```

```python
#checking na vales
dataset.isnull().sum()
```

```python
#replacing na with their mean values
dataset['Age'].fillna((dataset['Age'].mean().astype(int)),inplace=True)
```

```python
dataset['Cabin'].isnull().sum()/len(dataset)
```

```python
dataset.isnull().sum()
```

```python
dataset.drop(['Cabin'],axis=1,inplace=True)
dataset.dropna(inplace=True)
```
